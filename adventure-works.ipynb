```python
#importando as bibliotecas

import pandas as pd
import matplotlib.pyplot as plt
import pandera as pa
plt.style.use("seaborn")
```

```python
#importando os datasets com tratamento de missing
values= ['-','nan']
df = pd.read_csv('metal_bands_2017.csv', encoding= 'unicode_escape',na_values=values)
df.drop('Unnamed: 0',axis=1, inplace=True)
df.drop_duplicates(inplace=True)
```

```python
#Faz a validação dos dados

schema = pa.DataFrameSchema(
    columns = {
        "band_name": pa.Column(pa.String),
        "fans": pa.Column(pa.Int),
        "formed": pa.Column(pa.Float, nullable=True),
        "origin": pa.Column(pa.String, nullable=True),
        "split": pa.Column(pa.Float, nullable=True),
        "style": pa.Column(pa.String)
    }
)

schema.validate(df)
```

```python
#mostra a quantidade de linhas e colunas
df.shape
```

```python
#Filtro de bandas brasileiras
brasileiras = df.loc[df['origin']=='Brazil']
```

```python
#Quantidade de bandas de metal no Brasil
brasileiras.band_name.count()
```

```python
#Bandas de metal brasileiras ordenadas pela quantidade de fãs
brasileiras.sort_values(by='fans', ascending=False)
```

```python
#Quantidades de bandas de metal ativas no brasil
brasileiras.split.isna().sum()
```

```python
#bandas brasileiras agrupadas por ano de criação
brasileiras.groupby(['formed'])['band_name'].unique()
```

```python
#Países com mais bandas de metal
df.groupby(['origin'])['band_name'].count().sort_values(ascending=False).head()
```
